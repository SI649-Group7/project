```python
import numpy as np
import pandas as pd
import altair as alt
import warnings
warnings.filterwarnings('ignore')
```

```python
df_ms = pd.read_csv('data/ms.csv')
df_ms.sample(5)
```

```python
df_ms.columns
```

```python
df_battle = pd.DataFrame(columns=['catagory', 'player_1', 'player_2', 'name_1', 'name_2'])
names = sorted(np.unique(df_ms['team_one_players'].to_list() + df_ms['team_two_players'].to_list()))
len(names)
```

```python
win_rate = {}
for player_1 in names:
    win_rate[player_1] = {}
    for player_2 in names:
        if player_1 != player_2:
            win_rate[player_1][player_2] = {
                'win': 0,
                'total': 0,
            }
print(win_rate['Dan Lin'])
```

```python
for index, row in df_ms.iterrows():
    player_1 = row['team_one_players']
    player_2 = row['team_two_players']
    if row['winner'] == 1:
        win_rate[player_1][player_2]['win'] += 1
    elif row['winner'] == 2:
        win_rate[player_2][player_1]['win'] += 1
    win_rate[player_1][player_2]['total'] += 1
    win_rate[player_2][player_1]['total'] += 1
count = 0
for player in win_rate:
    for opponent in win_rate[player]:
        total = win_rate[player][opponent]['total']
        if total >= 3:
            count += 1
            win_1 = win_rate[player][opponent]['win']
            print(f'{player} vs {opponent}: {win_1} - {total - win_1}')
print(count)
```

```python
for player in win_rate:
    for opponent in win_rate[player]:
        total = win_rate[player][opponent]['total']
        win_1 = win_rate[player][opponent]['win']
        if total >= 3:
            df_battle = df_battle.append({
                'catagory': 'Winning Rate',
                'player_1': win_1/total,
                'player_2': (total - win_1)/total,
                'name_1': player,
                'name_2': opponent
            }, ignore_index=True)

df_battle
```

```python
player_1 = list(df_battle["name_1"].unique())
player_2 = list(df_battle["name_2"].unique())
player_1.sort()
player_2.sort()

dropdown1 = alt.binding_select(options=player_1, labels=player_1, name="player 1")
selection1 = alt.selection_single(fields=['name_1'], bind=dropdown1, name='player 1')

dropdown2 = alt.binding_select(options=player_2, labels=player_2, name="player 2")
selection2 = alt.selection_single(fields=['name_2'], bind=dropdown2, name='player 2')

op_condition = {"condition":
                {"test":
                 {"and":
                  [{"selection": "player 1"},
                   {"selection": "player 2"}]},
                    "value": 1},
                "value": 0}

base = alt.Chart(df_battle).properties(
    height=50,
    width=300
)

left = base.mark_bar(color='#1f77b4').encode(
    x=alt.X('player_1', sort=alt.SortOrder('descending'), axis=alt.Axis(format='%')),
    y=alt.Y('catagory', axis=None),
    opacity=op_condition
)

middle = base.encode(
    y=alt.Y('catagory:N', axis=None),
    text=alt.Text('catagory:N'),
).mark_text().properties(
    width=80
)

right = base.mark_bar(color='orange').encode(
    x=alt.X('player_2', axis=alt.Axis(format='%')),
    y=alt.Y('catagory', axis=None),
    opacity=op_condition
)

alt.concat(left, right, spacing=5).add_selection(
    selection2, selection1
).configure_scale(
    bandPaddingInner=0.2
).properties(
    title='Recent Years Winning Rate Comparison'
).configure_title(
    fontSize=20,
    anchor='middle'
)

```
